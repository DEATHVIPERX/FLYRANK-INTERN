## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Load the starter dataset
df = pd.read_csv('flyrank-ml-internship-starter/data/raw/content_refresh_anonymized.csv')

# 1. Total volume of our unit of analysis
total_pages = len(df)

# 2. Identify the class balance for our proxy label (downward trend)
declining_pages = len(df[df['trend_direction'] == 'down'])
decline_rate = (declining_pages / total_pages) * 100

# 3. Filter for actionable volume (pages with enough impressions to matter)
high_volume_declines = len(df[(df['trend_direction'] == 'down') & (df['impressions_90d'] >= 500)])

print(f"Total pages in starter slice: {total_pages:,}")
print(f"Pages showing a downward trend: {declining_pages:,} ({decline_rate:.1f}%)")
print(f"Actionable declining pages (>= 500 impressions): {high_volume_declines:,}")
```

What the data tells us: We aren't looking for a needle in a haystack; we are triaging a massive queue. With thousands of pages showing a downward trend, a simple rule-based approach will flag far too many pages for a human team to review. Filtering for actionable volume (e.g., >= 500 impressions) helps, but we still need a predictive model to rank these candidates by actual recovery probability and separate the true signal from the noise.
